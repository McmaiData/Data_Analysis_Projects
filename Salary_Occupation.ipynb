# Mean-Variance-Standard Deviation Calculator

Descrpcion del proyecto:
Dada una lista debes convertirla en un arreglo y obtener la media, varianza y ds
de las columnas y de las filas.


```python
import numpy as np
```

```python
def calculate(lista):
  m = np.array(lista).reshape(3,3)
  # Para la media
  media_c1 = np.mean(m[:,0])
  media_c2 = np.mean(m[:,1])
  media_c3 = np.mean(m[:,2])

  media_r1 = np.mean(m[0])
  media_r2 = np.mean(m[1])
  media_r3 = np.mean(m[2])

  media_total = np.mean(m)

  # Varianza
  varianza_c1 = np.var(m[:,0])
  varianza_c2 = np.var(m[:,1])
  varianza_c3 = np.var(m[:,2])

  varianza_r1 = np.var(m[0])
  varianza_r2 = np.var(m[1])
  varianza_r3 = np.var(m[2])

  varianza_total = np.var(m)

  # SD

  desviacion_c1 = np.std(m[:,0])
  desviacion_c2 = np.std(m[:,1])
  desviacion_c3 = np.std(m[:,2])

  desviacion_r1 = np.std(m[0])
  desviacion_r2 = np.std(m[1])
  desviacion_r3 = np.std(m[2])

  desviacion_total = np.std(m)

  # maximo
  maximo_c1 = np.max(m[:,0])
  maximo_c2 = np.max(m[:,1])
  maximo_c3 = np.max(m[:,2])
  maximo_r1 = np.max(m[0])
  maximo_r2 = np.max(m[1])
  maximo_r3 = np.max(m[2])
  maximo_total = np.max(m)

  # minimo
  minimo_c1 = np.min(m[:,0])
  minimo_c2 = np.min(m[:,1])
  minimo_c3 = np.min(m[:,2])
  minimo_r1 = np.min(m[0])
  minimo_r2 = np.min(m[1])
  minimo_r3 = np.min(m[2])
  minimo_total = np.min(m)

  # suma
  suma_c1 = np.sum(m[:,0])
  suma_c2 = np.sum(m[:,1])
  suma_c3 = np.sum(m[:,2])
  suma_r1 = np.sum(m[0])
  suma_r2 = np.sum(m[1])
  suma_r3 = np.sum(m[2])
  suma_total = np.sum(m)

  calculation = {
  'mean': [[media_c1.item(), media_c2.item(), media_c3.item()], [media_r1.item(),media_r2.item(),media_r3.item()], media_total.item()],
  'variance': [[varianza_c1.item(),varianza_c2.item(),varianza_c3.item()], [varianza_r1.item(),varianza_r2.item(),varianza_r3.item()], varianza_total.item()],
  'standard deviation': [[desviacion_c1.item(),desviacion_c2.item(),desviacion_c3.item()], [desviacion_r1.item(),desviacion_r2.item(),desviacion_r3.item()], desviacion_total.item()],
  'max': [[maximo_c1.item(),maximo_c2.item(),maximo_c3.item()], [maximo_r1.item(),maximo_r2.item(),maximo_r3.item()], maximo_total.item()],
  'min': [[minimo_c1.item(),minimo_c2.item(),minimo_c3.item()], [minimo_r1.item(),minimo_r2.item(),minimo_r3.item()], minimo_total.item()],
  'sum': [[suma_c1.item(),suma_c2.item(),suma_c3.item()], [suma_r1.item(),suma_r2.item(),suma_r3.item()], suma_total.item()]}

  return calculation
```

```python
calculate([9,1,5,3,3,3,2,9,0])
```

## Mejorado por AI

```python
import numpy as np

def calculate(lista):
    if len(lista) != 9:
        raise ValueError("La lista debe contener exactamente 9 elementos.")

    m = np.array(lista).reshape(3, 3)

    stats = {
        'mean': [np.mean(m, axis=0).tolist(), np.mean(m, axis=1).tolist(), np.mean(m).item()],
        'variance': [np.var(m, axis=0).tolist(), np.var(m, axis=1).tolist(), np.var(m).item()],
        'standard deviation': [np.std(m, axis=0).tolist(), np.std(m, axis=1).tolist(), np.std(m).item()],
        'max': [np.max(m, axis=0).tolist(), np.max(m, axis=1).tolist(), np.max(m).item()],
        'min': [np.min(m, axis=0).tolist(), np.min(m, axis=1).tolist(), np.min(m).item()],
        'sum': [np.sum(m, axis=0).tolist(), np.sum(m, axis=1).tolist(), np.sum(m).item()]
    }

    return stats
```

```python
calculate([9,1,5,3,3,3,2,9,0])
```
